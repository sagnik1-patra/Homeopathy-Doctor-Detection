```python
import os
import json
import yaml
import joblib
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# ===================================================================
# PATHS
# ===================================================================
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Homeopathy Doctor Detection\registration_of_indian_medicine_and_homoeopathy_practitioners_2007-08_2009.csv"
SAVE_PATH = r"C:\Users\NXTWAVE\Downloads\Homeopathy Doctor Detection"

os.makedirs(SAVE_PATH, exist_ok=True)

# ===================================================================
# LOAD DATA
# ===================================================================
df = pd.read_csv(DATA_PATH)
df = df.fillna(0)     # FIX — important

df_numeric = df.select_dtypes(include=[np.number])

X = df_numeric.iloc[:, :-1]
y = df_numeric.iloc[:, -1]

# ===================================================================
# SCALING
# ===================================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
test_size = 0.2 if len(df) > 5 else 0.0
if test_size == 0:
    X_train, y_train = X_scaled, y
    X_test, y_test = X_scaled, y
else:
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=test_size, random_state=42
    )

# ===================================================================
# BUILD SIMPLE MODEL
# ===================================================================
model = Sequential([
    Dense(32, activation="relu", input_shape=(X_train.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")

history = model.fit(X_train, y_train, epochs=30, batch_size=4, verbose=1)

# ===================================================================
# PREDICTION
# ===================================================================
pred_train = model.predict(X_train).flatten()
pred_test = model.predict(X_test).flatten()

# ===================================================================
# SAVE PREDICTION JSON
# ===================================================================
prediction_json_path = os.path.join(SAVE_PATH, "predictions_homeopathy.json")
with open(prediction_json_path, "w") as f:
    json.dump({"train_predictions": pred_train.tolist(),
               "test_predictions": pred_test.tolist()}, f, indent=4)

print("✔ Saved predictions JSON at:", prediction_json_path)

# ===================================================================
# SAVE RESULT CSV
# ===================================================================
result_df = pd.DataFrame({
    "Actual": y_train,
    "Predicted": pred_train
})

result_csv_path = os.path.join(SAVE_PATH, "results_homeopathy.csv")
result_df.to_csv(result_csv_path, index=False)

print("✔ Saved results CSV at:", result_csv_path)

# ===================================================================
# 📌 HEATMAP
# ===================================================================
plt.figure(figsize=(8, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()

heatmap_path = os.path.join(SAVE_PATH, "heatmap.png")
plt.savefig(heatmap_path)
plt.show()

print("✔ Saved heatmap at:", heatmap_path)

# ===================================================================
# 📌 ACCURACY GRAPH (LOSS CURVE)
# ===================================================================
plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Model Loss Curve")
plt.legend()
plt.tight_layout()

accuracy_graph_path = os.path.join(SAVE_PATH, "accuracy_graph.png")
plt.savefig(accuracy_graph_path)
plt.show()

print("✔ Saved accuracy graph at:", accuracy_graph_path)

# ===================================================================
# 📌 COMPARISON GRAPH (Actual vs Predicted)
# ===================================================================
plt.figure(figsize=(8, 6))
plt.plot(y_train.values, label="Actual", marker="o")
plt.plot(pred_train, label="Predicted", marker="x")
plt.xlabel("Samples")
plt.ylabel("Value")
plt.title("Actual vs Predicted Comparison")
plt.legend()
plt.tight_layout()

comparison_graph_path = os.path.join(SAVE_PATH, "comparison_graph.png")
plt.savefig(comparison_graph_path)
plt.show()

print("✔ Saved comparison graph at:", comparison_graph_path)

# ===================================================================
# 📌 PREDICTION GRAPH
# ===================================================================
plt.figure(figsize=(8, 6))
plt.plot(pred_train, label="Predictions", color="purple")
plt.xlabel("Samples")
plt.ylabel("Predicted Value")
plt.title("Prediction Graph")
plt.legend()
plt.tight_layout()

prediction_graph_path = os.path.join(SAVE_PATH, "prediction_graph.png")
plt.savefig(prediction_graph_path)
plt.show()

print("✔ Saved prediction graph at:", prediction_graph_path)

# ===================================================================
# 📌 RESULT GRAPH (Scatter + Trendline)
# ===================================================================
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_train, y=pred_train)
sns.regplot(x=y_train, y=pred_train, scatter=False, color="red")

plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.title("Result Graph (Actual vs Predicted)")
plt.tight_layout()

result_graph_path = os.path.join(SAVE_PATH, "result_graph.png")
plt.savefig(result_graph_path)
plt.show()

print("✔ Saved result graph at:", result_graph_path)

print("\n🎉 ALL GRAPHS + JSON + CSV SAVED SUCCESSFULLY!")
print("Location:", SAVE_PATH)
```
