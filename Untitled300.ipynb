```python
import os
import json
import yaml
import joblib
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


# ===============================================================
# PATHS
# ===============================================================
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Homeopathy Doctor Detection\registration_of_indian_medicine_and_homoeopathy_practitioners_2007-08_2009.csv"
SAVE_PATH = r"C:\Users\NXTWAVE\Downloads\Homeopathy Doctor Detection"
os.makedirs(SAVE_PATH, exist_ok=True)

# ===============================================================
# LOAD + CLEAN DATA
# ===============================================================
df = pd.read_csv(DATA_PATH)
df = df.fillna(0)

df_numeric = df.select_dtypes(include=[np.number])

X = df_numeric.iloc[:, :-1]
y = df_numeric.iloc[:, -1]

# scaling
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# split
test_size = 0.2 if len(df) > 5 else 0.0
if test_size == 0:
    X_train, y_train = X_scaled, y
    X_test, y_test = X_scaled, y
else:
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=test_size, random_state=42
    )


# ===============================================================
# HYBRID AI SYSTEM (AIS + PSO)
# ===============================================================

def create_model(units1, units2, lr):
    model = Sequential()
    model.add(Dense(units1, activation="relu", input_shape=(X_train.shape[1],)))
    model.add(Dense(units2, activation="relu"))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=lr), loss="mse")
    return model


def evaluate_candidate(candidate):
    """Train very fast (2 epochs) to evaluate fitness"""
    model = create_model(candidate["u1"], candidate["u2"], candidate["lr"])
    history = model.fit(X_train, y_train, epochs=2, batch_size=4, verbose=0)
    loss = history.history["loss"][-1]
    return loss


# initial population (AIS clone-based diversity)
def generate_population(n=5):
    pop = []
    for _ in range(n):
        c = {
            "u1": random.randint(8, 64),
            "u2": random.randint(8, 64),
            "lr": random.uniform(0.0005, 0.01),
            "velocity": {"u1": 0, "u2": 0, "lr": 0}
        }
        pop.append(c)
    return pop


# PSO update
def update_particle(particle, best_global):
    w, c1, c2 = 0.5, 1.5, 1.5

    for key in ["u1", "u2"]:
        r1, r2 = random.random(), random.random()
        particle["velocity"][key] = (
            w * particle["velocity"][key]
            + c1 * r1 * (best_global[key] - particle[key])
            + c2 * r2 * (best_global[key] - particle[key])
        )
        particle[key] += int(particle["velocity"][key])

        particle[key] = max(8, min(64, particle[key]))

    # update lr
    key = "lr"
    r1, r2 = random.random(), random.random()
    particle["velocity"][key] = (
        w * particle["velocity"][key]
        + c1 * r1 * (best_global[key] - particle[key])
        + c2 * r2 * (best_global[key] - particle[key])
    )
    particle[key] += particle["velocity"][key]
    particle[key] = max(0.0005, min(0.01, particle[key]))


# ======================================================
# RUN HYBRID AIS + PSO OPTIMIZATION
# ======================================================
population = generate_population()
best_global = None
best_loss = float("inf")

print("\nðŸ”µ Running AIS + PSO Hybrid Optimization...\n")

for iteration in range(6):
    print(f"[HYBRID] Iteration {iteration+1}/6")

    new_population = []
    for particle in population:
        loss = evaluate_candidate(particle)
        if loss < best_loss:
            best_loss = loss
            best_global = particle.copy()

        new_population.append(particle)

    # PSO update
    for particle in new_population:
        update_particle(particle, best_global)

    # AIS cloning (create slight random variations)
    clones = []
    for c in new_population:
        clone = c.copy()
        clone["u1"] += random.randint(-4, 4)
        clone["u2"] += random.randint(-4, 4)
        clone["lr"] += random.uniform(-0.001, 0.001)

        clone["u1"] = max(8, min(64, clone["u1"]))
        clone["u2"] = max(8, min(64, clone["u2"]))
        clone["lr"] = max(0.0005, min(0.01, clone["lr"]))
        clones.append(clone)

    population = new_population + clones


print("\nðŸŽ‰ BEST PARAMETERS FOUND:")
print(best_global)

# ===============================================================
# TRAIN FINAL MODEL USING BEST AIS+PSO PARAMS
# ===============================================================
final_model = create_model(best_global["u1"], best_global["u2"], best_global["lr"])

history = final_model.fit(X_train, y_train, epochs=30, batch_size=4, verbose=1)

pred_train = final_model.predict(X_train).flatten()
pred_test = final_model.predict(X_test).flatten()

# ===============================================================
# SAVE PREDICTIONS JSON
# ===============================================================
pred_json = os.path.join(SAVE_PATH, "hybrid_predictions_homeopathy.json")
with open(pred_json, "w") as f:
    json.dump({"train_predictions": pred_train.tolist(),
               "test_predictions": pred_test.tolist()}, f, indent=4)

print("âœ” Saved:", pred_json)

# ===============================================================
# SAVE RESULTS CSV
# ===============================================================
result_df = pd.DataFrame({"Actual": y_train, "Predicted": pred_train})
result_csv = os.path.join(SAVE_PATH, "hybrid_results_homeopathy.csv")
result_df.to_csv(result_csv, index=False)

print("âœ” Saved:", result_csv)

# ===============================================================
# ðŸ“Œ HEATMAP
# ===============================================================
plt.figure(figsize=(8, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Hybrid AIS+PSO Heatmap")
plt.tight_layout()

path = os.path.join(SAVE_PATH, "hybrid_heatmap.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)

# ===============================================================
# ðŸ“Œ ACCURACY GRAPH
# ===============================================================
plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.title("Hybrid AIS+PSO Accuracy Graph (Loss Curve)")
plt.xlabel("Epochs"); plt.ylabel("Loss")
plt.legend(); plt.tight_layout()

path = os.path.join(SAVE_PATH, "hybrid_accuracy_graph.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)

# ===============================================================
# ðŸ“Œ COMPARISON GRAPH
# ===============================================================
plt.figure(figsize=(8, 6))
plt.plot(y_train.values, label="Actual", marker="o")
plt.plot(pred_train, label="Predicted", marker="x")
plt.title("Hybrid AIS+PSO - Actual vs Predicted")
plt.xlabel("Samples"); plt.ylabel("Value")
plt.legend(); plt.tight_layout()

path = os.path.join(SAVE_PATH, "hybrid_comparison_graph.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)

# ===============================================================
# ðŸ“Œ PREDICTION GRAPH
# ===============================================================
plt.figure(figsize=(8, 6))
plt.plot(pred_train, color="purple", label="Prediction")
plt.title("Hybrid AIS+PSO Prediction Trend")
plt.xlabel("Samples"); plt.ylabel("Predicted")
plt.legend(); plt.tight_layout()

path = os.path.join(SAVE_PATH, "hybrid_prediction_graph.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)

# ===============================================================
# ðŸ“Œ RESULT GRAPH
# ===============================================================
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_train, y=pred_train)
sns.regplot(x=y_train, y=pred_train, scatter=False, color="red")
plt.title("Hybrid AIS+PSO Result Graph")
plt.xlabel("Actual"); plt.ylabel("Predicted")
plt.tight_layout()

path = os.path.join(SAVE_PATH, "hybrid_result_graph.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)

print("\nðŸŽ‰ ALL HYBRID AIS+PSO FILES + GRAPHS SAVED SUCCESSFULLY!")
print("LOCATION:", SAVE_PATH)
```
