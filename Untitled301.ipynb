```python
import os
import json
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


# ===================================================================
# PATHS
# ===================================================================
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Homeopathy Doctor Detection\registration_of_indian_medicine_and_homoeopathy_practitioners_2007-08_2009.csv"
SAVE_PATH = r"C:\Users\NXTWAVE\Downloads\Homeopathy Doctor Detection"
os.makedirs(SAVE_PATH, exist_ok=True)

# ===================================================================
# LOAD + CLEAN DATA
# ===================================================================
df = pd.read_csv(DATA_PATH)
df = df.fillna(0)

df_numeric = df.select_dtypes(include=[np.number])
X = df_numeric.iloc[:, :-1]
y = df_numeric.iloc[:, -1]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

test_size = 0.2 if len(df) > 5 else 0.0
if test_size == 0:
    X_train, y_train = X_scaled, y
    X_test, y_test = X_scaled, y
else:
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=test_size, random_state=42
    )


# ===================================================================
# MODEL CREATION
# ===================================================================
def create_model(u1, u2, lr):
    model = Sequential()
    model.add(Dense(u1, activation="relu", input_shape=(X_train.shape[1],)))
    model.add(Dense(u2, activation="relu"))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=lr), loss="mse")
    return model


def evaluate_candidate(c):
    model = create_model(c["u1"], c["u2"], c["lr"])
    history = model.fit(X_train, y_train, epochs=2, batch_size=4, verbose=0)
    return history.history["loss"][-1]


# ===================================================================
# HYBRID AIS + CSA SYSTEM
# ===================================================================

def levy_flight(Lambda=1.5):
    u = np.random.randn() * 0.01
    v = np.random.randn()
    step = u / (abs(v) ** (1 / Lambda))
    return step


def mutate(candidate):
    clone = candidate.copy()
    clone["u1"] = max(8, min(64, clone["u1"] + random.randint(-5, 5)))
    clone["u2"] = max(8, min(64, clone["u2"] + random.randint(-5, 5)))
    clone["lr"] = max(0.0005, min(0.01, clone["lr"] + random.uniform(-0.001, 0.001)))
    return clone


def cuckoo_search(candidate):
    new_candidate = candidate.copy()
    new_candidate["u1"] += int(levy_flight() * 10)
    new_candidate["u2"] += int(levy_flight() * 10)
    new_candidate["lr"] += levy_flight() * 0.001

    new_candidate["u1"] = max(8, min(64, new_candidate["u1"]))
    new_candidate["u2"] = max(8, min(64, new_candidate["u2"]))
    new_candidate["lr"] = max(0.0005, min(0.01, new_candidate["lr"]))

    return new_candidate


# ===================================================================
# RUN AIS + CSA HYBRID OPTIMIZATION
# ===================================================================
population = []
for _ in range(5):
    population.append({
        "u1": random.randint(8, 64),
        "u2": random.randint(8, 64),
        "lr": random.uniform(0.0005, 0.01)
    })

best_solution = None
best_loss = float("inf")

print("\nðŸ”µ Running AIS + CSA Hybrid Optimization...\n")

for iteration in range(6):
    print(f"[ASA] Iteration {iteration+1}/6")

    new_pop = []

    for cand in population:
        loss = evaluate_candidate(cand)

        if loss < best_loss:
            best_loss = loss
            best_solution = cand.copy()

        # AIS clone & mutate
        mutated = mutate(cand)

        # CSA cuckoo operations
        cuckoo = cuckoo_search(cand)

        new_pop += [cand, mutated, cuckoo]

    # keep top 5 best
    scored = [(evaluate_candidate(c), c) for c in new_pop]
    scored.sort(key=lambda x: x[0])
    population = [c for _, c in scored[:5]]

print("\nðŸŽ‰ BEST AIS + CSA PARAMETERS FOUND:")
print(best_solution)


# ===================================================================
# TRAIN FINAL MODEL
# ===================================================================
final_model = create_model(best_solution["u1"], best_solution["u2"], best_solution["lr"])
history = final_model.fit(X_train, y_train, epochs=30, batch_size=4, verbose=1)

pred_train = final_model.predict(X_train).flatten()
pred_test = final_model.predict(X_test).flatten()

# ===================================================================
# SAVE JSON
# ===================================================================
json_path = os.path.join(SAVE_PATH, "asa_predictions_homeopathy.json")
with open(json_path, "w") as f:
    json.dump({"train_pred": pred_train.tolist(), "test_pred": pred_test.tolist()}, f, indent=4)

print("âœ” Saved:", json_path)

# ===================================================================
# SAVE RESULT CSV
# ===================================================================
result_df = pd.DataFrame({"Actual": y_train, "Predicted": pred_train})
csv_path = os.path.join(SAVE_PATH, "asa_results_homeopathy.csv")
result_df.to_csv(csv_path, index=False)
print("âœ” Saved:", csv_path)


# ===================================================================
# ðŸ”¥ GRAPH 1 â€” HEATMAP
# ===================================================================
plt.figure(figsize=(8, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("ASA Heatmap")
plt.tight_layout()

path = os.path.join(SAVE_PATH, "asa_heatmap.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)


# ===================================================================
# ðŸ”¥ GRAPH 2 â€” ACCURACY / LOSS GRAPH
# ===================================================================
plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.title("ASA Loss Curve")
plt.xlabel("Epochs"); plt.ylabel("Loss")
plt.legend(); plt.tight_layout()

path = os.path.join(SAVE_PATH, "asa_accuracy_graph.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)


# ===================================================================
# ðŸ”¥ GRAPH 3 â€” COMPARISON GRAPH
# ===================================================================
plt.figure(figsize=(8, 6))
plt.plot(y_train.values, marker="o", label="Actual")
plt.plot(pred_train, marker="x", label="Predicted")
plt.title("ASA - Actual vs Predicted")
plt.legend(); plt.tight_layout()

path = os.path.join(SAVE_PATH, "asa_comparison_graph.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)


# ===================================================================
# ðŸ”¥ GRAPH 4 â€” PREDICTION GRAPH
# ===================================================================
plt.figure(figsize=(8, 6))
plt.plot(pred_train, color="purple", label="Prediction Trend")
plt.title("ASA Prediction Graph")
plt.legend(); plt.tight_layout()

path = os.path.join(SAVE_PATH, "asa_prediction_graph.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)


# ===================================================================
# ðŸ”¥ GRAPH 5 â€” RESULT GRAPH
# ===================================================================
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_train, y=pred_train)
sns.regplot(x=y_train, y=pred_train, scatter=False, color="red")
plt.title("ASA Result Graph")
plt.tight_layout()

path = os.path.join(SAVE_PATH, "asa_result_graph.png")
plt.savefig(path); plt.show()
print("âœ” Saved:", path)


print("\nðŸŽ‰ ALL AIS + CSA (ASA) FILES GENERATED SUCCESSFULLY!")
print("Saved in:", SAVE_PATH)
```
